# 2.感知机

感知机是一种非常简单的二分类线性分类模型，同时也是神经网络和支持向量机的基础。

感知机接收多个输入信号，输出一个信号。

```python
# 感知机会计算输入信号和权重的乘积，然后加上偏置，如果这个值大于0则输出1，否则输出0
def AND(x1, x2: int) -> int:
    w1, w2, b = 0.5, 0.5, -0.7  # 权重和偏置
    tmp = x1 * w1 + x2 * w2 + b  # 计算加权和
    if tmp <= 0:
        return 0
    else:
        return 1


print(AND(0, 0))  # 0
print(AND(0, 1))  # 0
print(AND(1, 0))  # 0
print(AND(1, 1))  # 1
```

换成 numpy 实现

```python
import numpy as np


def AND(x1, x2: int) -> int:
    x = np.array([x1, x2])
    w = np.array([0.5, 0.5])
    b = -0.7
    tmp = np.sum(w * x) + b
    if tmp <= 0:
        return 0
    else:
        return 1
```

权重控制输入信号的重要性的参数，而偏置是调整神经元被激活的容易程度

```python
# NAND 门与 AND 门类似，但输出结果相反
def NAND(x1, x2: int) -> int:
    x = np.array([x1, x2])
    w = np.array([-0.5, -0.5]) # 仅权重和偏置与AND不同
    b = 0.7
    tmp = np.sum(w * x) + b
    if tmp <= 0:
        return 0
    else:
        return 1

# OR 门
def OR(x1, x2: int) -> int:
    x = np.array([x1, x2])
    w = np.array([0.5, 0.5]) # 仅权重和偏置与AND不同
    b = -0.2
    tmp = np.sum(w * x) + b
    if tmp <= 0:
        return 0
    else:
        return 1
```

前面的感知机无法实现异或（XOR）门，因为异或门的输出不是线性可分的。 单层感知机无法分离非线性空间。但是我们可以通过组合多个感知机来实现异或门。

```python
# 使用 2 层感知机可以表示异或门
def XOR(x1, x2: int) -> int:
    s1 = NAND(x1, x2)
    s2 = OR(x1, x2)
    y = AND(s1, s2)
    return y
```
