```python
import torch
import torch.nn as nn
```

```python
# 输入数据
input = torch.randn(3, 2)
print(input)
# 得到预测概率值
pred = torch.sigmoid(input)
print(pred)
```

```python
# 目标值
target = torch.tensor([[0,1],[1,0],[0,1]], dtype=torch.float)
print(target)
```

```python
# 定义损失函数 二元交叉熵损失函数（常用于二分类问题）
loss = nn.BCELoss()
print(loss(pred, target)) # 预测概率值必须是 0 到 1 之间
```
