```python
import torch
import torch.nn as nn
```

```python
# 定义一个全连接层
linear = nn.Linear(5, 2)
```

```python
# 1. 常数初始化
nn.init.zeros_(linear.weight)
print(linear.weight)
```

```python
nn.init.ones_(linear.weight)
print(linear.weight)
```

```python
nn.init.constant_(linear.weight, 10)
print(linear.weight)
```

```python
# 2.秩初始化
nn.init.eye_(linear.weight)
print(linear.weight)
```

```python
# 3.正态分布初始化
nn.init.normal_(linear.weight, mean=0, std=1)
print(linear.weight)
```

```python
# 均匀分布初始化
nn.init.uniform_(linear.weight, a=-3, b=3)
print(linear.weight)
```

```python
# 5.Xavier 初始化 根据输入输出神经元数量调整权重的初始范围，确保每层输出方差与输入方差相近。
# 适用于 sigmoid 和 Tanh 等激活函数，有效缓解梯度消失或梯度爆炸问题
nn.init.xavier_normal_(linear.weight)
print(linear.weight)
```

```python
nn.init.xavier_uniform_(linear.weight)
print(linear.weight)
```

```python
# 6. He 初始化 适用于 relu 和 relu 的变体
nn.init.kaiming_normal_(linear.weight)
print(linear.weight)
```

```python
nn.init.kaiming_uniform_(linear.weight)
print(linear.weight)
```
