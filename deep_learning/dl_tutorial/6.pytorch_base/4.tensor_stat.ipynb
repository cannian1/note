```python
import torch
```

```python
tensor1 = torch.randint(1, 10, (3, 2, 4)).float()
print(tensor1)
```

```python
# sum 求和
print(tensor1.sum()) # 所有数求总和
```

```python
print(tensor1.sum(dim=2))  # 将指定维度消去，聚合求和
```

```python
# mean 求均值
print(tensor1.mean())
```

```python
print(tensor1.mean(dim=2))
```

```python
# std 求标准方差
print(tensor1.std())
```

```python
# max、min 求最大最小值
print(tensor1.max())
```

```python
print(tensor1.min(dim=2))  # 得到两个元组，values 是指定维度值的最小值，indices 是最小值的索引
```

```python
# 直接获取最大、最小值的索引
print(tensor1.argmin(dim=2))
```

```python
print(tensor1.argmin()) # 向量化展开获取最小索引
```

```python
# unique 去重
print(tensor1.unique())  # 展开为一维张量并去重
```

```python
# 排序
print(tensor1.sort()) # 默认按最后一个维度排序
```
