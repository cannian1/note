四则运算

- +、-、*、/加减乘除
- add()、sub()、mul()、div() 加减乘除，不改变原数据
- add_()、sub_()、mul_()、div_() 加减乘除、修改原数据

```python
tensor1 = torch.randint(low=0, high=10, size=(2, 3))
print(tensor1)
```

```python
print(tensor1 + 10)  # 广播每个元素相加，不修改原数据

tensor1.add(10)  # 广播每个元素相加，不修改原数据
print(tensor1)
```

```python
tensor1.add_(10)  # 会修改原数据
print(tensor1)
```

计算张量的哈达玛积

```python
# 计算张量的哈达玛积
tensor1 = torch.tensor([[1, 2], [3, 4]])
tensor2 = torch.tensor([[1, 2], [3, 4]])
print(tensor1 + tensor2)
print(tensor1 * tensor2)
```

```python
print(tensor1.mul(tensor2))
```

矩阵乘法运算

mm()严格用于二维矩阵相乘。

@、matmul()支持多维张量，按最后两个维度做矩阵乘法，其他维度相同，或者至少一个张量对应维度为1，广播后进行运算。

```python
# 2维矩阵的矩阵乘法
tensor1 = torch.tensor([[1, 2, 3], [4, 5, 6]])
tensor2 = torch.tensor([[1, 2], [3, 4], [5, 6]])

print(tensor1 @ tensor2)
```

```python
print(tensor1.matmul(tensor2))
```

```python
print(tensor1.mm(tensor2))
```

```python
# 3维张量的矩阵乘法
tensor1 = torch.tensor([[[1, 2, 3], [4, 5, 6]], [[6, 5, 4], [3, 2, 1]]])  # 2 * 2 * 3
tensor2 = torch.tensor([[[1, 2], [3, 4], [5, 6]], [[6, 5], [4, 3], [2, 1]]]) # 2 * 3 * 2

# 取 tensor1 最后两个维度 2*3 和 tensor2 最后两个维度 3*2 做矩阵乘法
print(tensor1 @ tensor2)
```

```python
print(tensor1.matmul(tensor2))
```
