```python
import torch
```

```python
tensor1 = torch.randint(1, 10, (3, 5, 4))
print(tensor1)
```

```python
# 1. 简单索引
print(tensor1[2, 1, 3])
```

```python
print(tensor1[2, 3])
```

```python
print(tensor1[1])
```

```python
print(tensor1[:, 3])  # 所有矩阵第3行
```

```python
# 2. 范围索引
print(tensor1[1:])  # 输出后两个矩阵
```

```python
print(tensor1[-1, 1:4, 0:3])
```

```python
print(tensor1[-1, 1:4:2, 0:3:2])
```

```python
# 列表索引
print(tensor1[
          [0, 1],
          [3, 2]])  # 每个维度的索引构成一个列表
```

```python
print(tensor1[
          [0, 1],
          [3, 2],
          [2, 3]])  # 每个维度的索引构成一个列表
# [第 0 维 = 0，第1维 = 3，第2维=2, 第 0 维 = 1，第1维 = 2，第2维=3]
```

```python
print(tensor1[
          [[0], [1]],
          [[1, 2]]])

# 2 * 2 * 4
# [[[第0维=0 第1维=1],
# [第0维=0 第1维=2]],
# [[第0维=1 第1维=1],
# [第0维=1 第1维=2]]]
```

```python
# 布尔索引
# 选取符合条件的行，首元素大于5
mask = tensor1[:, :, 0] > 5
print(mask)  # 3 * 5
```

```python
print(tensor1[mask])  # 过滤出符合条件的数据
```

```python
# 选取符合条件的列，第二个元素大于5
mask = tensor1[:, 1, :] > 5
print(mask)  # 3 * 4
```

```python
tensor2 = tensor1.mT  # mT 是对矩阵转置，直接用 T 是把 3 * 5 * 4 转变为 4*5*3
print(tensor2)
```

```python
print(tensor2[mask].mT)
```

```python
# 选取符合条件的矩阵
# 元素 (1,2) >5
mask = tensor1[:, 1, 2] > 5
print(mask)
```

```python
print(tensor1[mask])
```
