```python
import torch
```

```python
x = torch.tensor(2.0, requires_grad=True)
y = x.detach()
```

```python
print(x.requires_grad)
print(y.requires_grad)  # False
```

```python
# 分离后不是同一个对象
print(id(x))
print(id(y))
```

```python
# 底层数据存储是共享的
print(x.untyped_storage().data_ptr())
print(y.untyped_storage().data_ptr())
```

```python
z1 = x ** 2
z2 = y ** 2
print(z1)
print(z2)
```

```python
z1.sum().backward()
# z2.sum().backward() # 没有启用梯度计算，调用会报错
```

```python
x = torch.ones(2, 2, requires_grad=True)
y = x * x
print(x)
print(y)
```

```python
u = y.detach()
print(u)
```

```python
# 让 u 参与新张量的运算
z = u * x

# 反向传播，计算所有梯度
z.sum().backward()
print(x.grad)

print(x.grad == u)
```
