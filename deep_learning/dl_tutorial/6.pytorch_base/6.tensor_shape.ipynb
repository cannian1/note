```python
import torch
```

```python
tensor1 = torch.randint(1, 10, (2, 3, 6))
print(tensor1)
```

交换维度

```python
# 交换两个维度
print(tensor1.transpose(1, 2))  # 第 1 维和 第 2 维 交换
```

```python
# 重新排列多个维度
print(tensor1.permute(0, 2, 1))
```

调整形状

```python
# 1.reshape
tensor1 = torch.randint(1, 10, (3, 5, 4))
print(tensor1.shape)
```

```python
print(tensor1.reshape(6, 10))  # 相乘的总元素个数要和原始张量的总元素个数相同
```

```python
print(tensor1.reshape(3, -1))  # 第二个参数自动计算
```

```python
print(tensor1.reshape(-1))  # 行向量化
```

```python
# 2. 用 view 调整
print(tensor1.is_contiguous())  # view 只能在连续的内存空间上使用，所以要判断是否内存空间上连续
```

```python
print(tensor1.view(-1, 10))
```

```python
print(tensor1.shape)
```

```python
# 对张量进行转置操作，让内存空间不连续
tensor1 = tensor1.permute(2, 1, 0)
print(tensor1.shape)
print(tensor1.is_contiguous())
```

```python
print(tensor1.contiguous().view(-1, 10))  # 强制内存连续后才能用 view 调整
```

增加或删除维度

```python
tensor1 = torch.tensor([1, 2, 3])
print(tensor1)
```

```python
# 在 0 维上增加一个维度, 1*3
print(tensor1.unsqueeze(dim=0))
```

```python
# 在 1 维上增加一个维度, 3*1
print(tensor1.unsqueeze(dim=1))
print(tensor1)
```

```python
# 在-2维上增加一个维度, 原地操作
print(tensor1.unsqueeze_(dim=-2))
print(tensor1)
```

```python
# 减少维度
print(tensor1.squeeze(dim=0))
```

张量拼接操作

```python
# 拼接
tensor1 = torch.randint(1, 10, (2, 2, 5))  # 2 * 2 * 5
tensor2 = torch.randint(1, 10, (2, 3, 5))  # 2 * 3 * 5
print(tensor1)
print(tensor2)
```

```python
# cat 拼接
# 按已有维度拼接。除拼接维度外，其他维度大小须相同
print(torch.cat((tensor1, tensor2), dim=1))  # 2 * 5 * 5
```

```python
# stack 张量堆叠，按新维度堆叠。所有张量形状必须一致。可以都放到 dim = 0，然后再调形状
tensor1 = torch.randint(1, 10, (4, 3, 5))  # 4 * 2 * 5
tensor2 = torch.randint(1, 10, (4, 3, 5))  # 4 * 3 * 5
print(tensor1)
print(tensor2)
```

```python
tensor3 = torch.stack([tensor1, tensor2], dim=0)
print(tensor3)
print(tensor3.shape)
```

```python
tensor3 = torch.stack([tensor1, tensor2], dim=1) # 增加的维度放到第一维
print(tensor3.shape)
```
