# 饼图



```python
import matplotlib.pyplot as plt
from matplotlib import rcParams  # 导入rcParams以设置字体

# 设置字体为中文
rcParams['font.sans-serif'] = ['SimHei']  # 设置字体为黑体

# 创建图表，设置大小
plt.figure(figsize=(10, 6))

# 数据
age_groups = ['0-14岁', '15-64岁', '65岁及以上']  # 各年龄段人口
population = [23063, 96228, 21676]  # 各年龄段人口数量

explode = (0.1, 0, 0.1)  # 突出显示0-14岁和65岁及以上的部分

# 设置饼图的颜色
colors = ['#ADD8E6', '#8DB6FF', '#90EE90']
# 绘制饼图
plt.pie(population,
        labels=age_groups,
        autopct='%1.1f%%',  # 设置百分比格式
        colors=colors,  # 设置饼图颜色
        startangle=140,  # 设置起始角度
        explode=explode,  # 突出显示部分
        shadow=True,  # 添加阴影效果
        )

plt.title('2023年中国各年龄段人口分布', fontsize=20)  # 设置标题

plt.tight_layout()  # 自动调整布局
plt.show()
```

```python
# 绘制环形图

import matplotlib.pyplot as plt
from matplotlib import rcParams  # 导入rcParams以设置字体

# 设置字体为中文
rcParams['font.sans-serif'] = ['SimHei']  # 设置字体为黑体

# 创建图表，设置大小
plt.figure(figsize=(10, 6))

# 数据
age_groups = ['0-14岁', '15-64岁', '65岁及以上']  # 各年龄段人口
population = [23063, 96228, 21676]  # 各年龄段人口数量

# 设置饼图的颜色
colors = ['#ADD8E6', '#8DB6FF', '#90EE90']
# 绘制饼图
'''
    wedgeprops={'width': 0.5} 设置环形图的宽度
    pctdistance=0.75 设置百分比标签距离圆心的距离
'''
plt.pie(population, labels=age_groups, autopct='%1.1f%%', colors=colors, startangle=140, wedgeprops={'width': 0.5},
        pctdistance=0.75)

plt.text(0, 0, '2023年', ha='center', va='center', fontsize=20, color='black')  # 在中心添加文本

plt.tight_layout()  # 自动调整布局
plt.show()
```
