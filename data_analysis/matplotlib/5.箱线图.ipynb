# 箱线图

箱线图（Box Plot）是一种用于显示数据分布的统计图表。它通过箱体和须来展示数据的中位数、四分位数以及异常值等信息。

箱体表示数据的中间50%的分布情况，箱体的上边缘和下边缘分别表示上四分位数（Q3）和下四分位数（Q1），箱体中间的线表示中位数（Q2）。箱体外的线段（须）表示数据的范围，通常延伸到1.5倍四分位距（IQR）之外的数据点。超出这个范围的数据点被视为异常值。

```python
import matplotlib.pyplot as plt
from matplotlib import rcParams # 导入rcParams以设置字体

# 设置字体为中文
rcParams['font.sans-serif'] = ['SimHei']  # 设置字体为黑体

data = {
    '语文': [55, 85, 90, 95, 100, 105, 110, 115, 120, 125],
    '数学': [82, 88, 92, 96, 100, 104, 108, 112, 116, 120],
    '英语': [42, 48, 52, 56, 60, 64, 68, 86, 90, 101],
}

plt.figure(figsize=(10, 6))

plt.boxplot(data.values(), tick_labels=data.keys())
plt.title('各科成绩箱线图')
plt.ylabel('分数')
plt.grid(True,axis='y', linestyle='--', alpha=0.7)

plt.show()
```
