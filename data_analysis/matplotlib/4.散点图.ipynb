# 散点图

```python
# 绘制折线图
import matplotlib.pyplot as plt
from matplotlib import rcParams # 导入rcParams以设置字体

# 设置字体为中文
rcParams['font.sans-serif'] = ['SimHei']  # 设置字体为黑体
rcParams['axes.unicode_minus'] = False  # 解决负号显示问题

# 创建图表，设置大小
plt.figure(figsize=(10, 6))

# 生成数据
x = []
y = []
for i in range(100):
    x.append(i)
    y.append(i**2 + (i % 10) * 10)

# 绘制散点图
# s 参数设置点的大小
# alpha 参数设置透明度
plt.scatter(x, y, color='blue', s=10, alpha=0.6)

# 参考线
plt.plot([0, 100], [0, 6000], color='orange', linestyle='-', label='y=x^2') # 添加参考线


plt.show()
```
