# 多个图表绘制

```python
# 绘制折线图
import matplotlib.pyplot as plt
from matplotlib import rcParams  # 导入rcParams以设置字体

# 设置字体为中文
rcParams['font.sans-serif'] = ['SimHei']  # 设置字体为黑体
rcParams['axes.unicode_minus'] = False  # 解决负号显示问题

plt.figure(figsize=(10, 6))


month = ['2024年\n7月', '8月', '9月', '10月', '11月', '12月', '2025年\n1-2月', '3月', '4月', '5月', '6月', '7月']

# 产量
yield_data = [284.9, 292.7, 267.5, 235.8, 249.8, 273.0, 252.9, 251.0, 237.0, 238.0, 265.4, 298.9]

# 增速
growth_rate = [2.5,5.8,6.0,2.1,0.9,0.6,-1.3,1.8,0.9,0.5,1.7,3.1]

f1 = plt.subplot(1, 2, 1)  # 创建第一个子图 行数=1，列数=2，位置=1
# 设置 y 轴范围
f1.set_ylim(100, 400)
f1.bar(month, yield_data, color='#8DB6FF', width=0.5, label='产量')
f1.set_ylabel('日均产量（亿千瓦时）')

f2 = plt.subplot(1, 2, 2)
# 设置 y 轴范围
f2.set_ylim(-50, 20)
f2.plot(month, growth_rate, marker='o', color='#FF8C00', label='产量')

plt.tight_layout()  # 自动调整子图参数，使之填充整个图像区域
plt.show()
```

```python
import matplotlib.pyplot as plt

# 示例数据

month = ['2024年\n7月', '8月', '9月', '10月', '11月', '12月', '2025年\n1-2月', '3月', '4月', '5月', '6月', '7月']
# 产量(柱状图数据)
yield_data = [284.9, 292.7, 267.5, 235.8, 249.8, 273.0, 252.9, 251.0, 237.0, 238.0, 265.4, 298.9]
# 增速(折线图数据)
growth_rate = [2.5,5.8,6.0,2.1,0.9,0.6,-1.3,1.8,0.9,0.5,1.7,3.1]

# 创建图形和第一个轴（柱状图）
fig, ax1 = plt.subplots(figsize=(8, 6))

# 绘制柱状图（左侧Y轴）
bars = ax1.bar(month, yield_data, color='skyblue', alpha=0.7, label='日均产量（亿千瓦时）')
ax1.set_ylabel('日均产量（亿千瓦时）', color='blue')
ax1.tick_params(axis='y', labelcolor='blue')
ax1.set_ylim(100, 400)  # 设置左侧Y轴范围

# 创建第二个Y轴（折线图）
ax2 = ax1.twinx()

# 绘制折线图（右侧Y轴）
line, = ax2.plot(month, growth_rate, color='red', marker='o', label='增速')
ax2.set_ylabel('增速', color='red')
ax2.tick_params(axis='y', labelcolor='red')
ax2.set_ylim(-50, 20)  # 设置右侧Y轴范围

# 添加图例（合并两个图的图例）
lines = [bars, line]
labels = [l.get_label() for l in lines]
ax1.legend(lines, labels, loc='upper left')

# 添加标题
plt.title('规模以上工业发电量同比增速及日均产量',pad=20)

# 显示图形
plt.show()
```
