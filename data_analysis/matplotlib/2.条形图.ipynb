# 条形图

条形图是用来显示分类数据的图表。它通过条形的长度或高度来表示不同类别的数值大小。


```python
# 绘制柱状图 （竖着）
import matplotlib.pyplot as plt
from matplotlib import rcParams  # 导入rcParams以设置字体
import textwrap  # 导入textwrap模块用于文本换行


# 定义自动换行函数（每4个字符换行）
def wrap_labels(labels, max_len=4):
    return ["\n".join(textwrap.wrap(label, max_len)) for label in labels]


# 设置字体为中文
rcParams['font.sans-serif'] = ['SimHei']  # 设置字体为黑体
plt.rcParams['axes.unicode_minus'] = False  # 解决负号显示为方块的问题

# 创建图表，设置大小
plt.figure(figsize=(10, 6))

# 数据
# https://www.stats.gov.cn/sj/zxfb/202508/t20250809_1960635.html
items = ['食品烟酒', '衣着', '居住', '生活用品及服务', '交通通信', '教育文化娱乐', '医疗保健', '其他用品及服务']
rate = [-0.1, -0.3, 0.1, 0.8, 1.5, 1.3, 0.2, 0.9]
colors = ['#FFB347' if r >= 0 else '#8DB6FF' for r in rate]

# 应用换行
wrapped_items = wrap_labels(items)

# 创建柱状图
plt.bar(wrapped_items, rate, width=0.4, color=colors, edgecolor='black', linewidth=0.7)
# 添加标题和标签
plt.title('7月份居民消费价格分类别环比涨跌幅', fontsize=16)
# 设置x轴和y轴标签
plt.xlabel('各类商品及服务价格同比变动情况', fontsize=14, labelpad=20)
plt.ylabel('环比(%)', fontsize=14)

# 显示网格
plt.grid(axis='y', linestyle='--', color='#EEEEEE', alpha=0.7)

# 在y=0的位置添加黑色参考线
plt.axhline(y=0, color='black', linewidth=1.2, linestyle='-')

# 设置刻度字体大小
plt.xticks(fontsize=12)
plt.yticks(fontsize=12)

# 设置 y 轴范围
plt.ylim(-0.5, 2.0)
plt.gca().yaxis.set_major_formatter(plt.FuncFormatter(lambda x, _: f'{x:.1f}'))  # 设置y轴格式化为一位小数

# 在每个条形上添加数据标签
for i, v in enumerate(rate):
    plt.text(i, v + 0.035 if v > 0 else v - 0.1, str(v), ha='center', va='bottom', fontsize=12)

#plt.tight_layout()  # 自动调整布局

plt.show()
```

```python
# 绘制条形图 （横着）
import matplotlib.pyplot as plt
from matplotlib import rcParams  # 导入rcParams以设置字体

# 设置字体为中文
rcParams['font.sans-serif'] = ['SimHei']  # 设置字体为黑体

# 创建图表，设置大小
plt.figure(figsize=(10, 6))

# 数据
weather = ['晴', '多云', '雨', '阴', '沙雾霾', '雪']
days = [237, 141, 89, 87, 37, 2]
colors = ['#FFD700', '#87CEEB', '#EE0000', '#D3D3D3', '#FF8C00', '#66CCFF']

# 将天气和天数组合成元组列表，并按天数排序
data = list(zip(weather, days, colors))
data.sort(key=lambda x: x[1])  # 按天数升序排序
weather_sorted = [item[0] for item in data]
days_sorted = [item[1] for item in data]
colors_sorted = [item[2] for item in data]

# 创建条形图
plt.barh(weather_sorted, days_sorted, color=colors_sorted)
# 添加标题和标签
plt.title('2024年的天气情况', fontsize=16)
# 设置x轴和y轴标签
plt.xlabel('天数', fontsize=14)
plt.ylabel('天气类型', fontsize=14)

# 显示网格
plt.grid(axis='x', linestyle='--', alpha=0.7)
# 设置刻度字体大小
plt.xticks(fontsize=12)
plt.yticks(fontsize=12)
# 设置 x 轴范围
plt.xlim(0, 250)
# 在每个条形上添加数据标签
for i, v in enumerate(days_sorted):
    plt.text(v + 3, i, str(v), ha='left', va='center', fontsize=12)

plt.tight_layout()  # 自动调整布局
plt.show()
```
