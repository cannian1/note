# 折线图 plot

折线图是用来显示数据随时间变化的趋势，通常用于展示连续数据的变化情况。它通过连接数据点来显示数据的趋势和波动。

```python
# 绘制折线图
import matplotlib.pyplot as plt
from matplotlib import rcParams # 导入rcParams以设置字体

# 设置字体为中文
rcParams['font.sans-serif'] = ['SimHei']  # 设置字体为黑体

# 创建图表，设置大小
plt.figure(figsize=(10, 6))

# 定义数据
months = ['一月', '二月', '三月', '四月', '五月', '六月', '七月', '八月', '九月', '十月', '十一月', '十二月']
sales = [150, 200,700, 300, 157, 400, 450, 500, 550, 600, 650, 700]

# 绘制折线图，设置线条颜色、样式和标记
# linestyle 可选值包括 '-'（实线）, '--'（虚线）, '-.'（点划线）, ':'（点线）
# marker 可选值包括 'o'（圆圈）, 's'（方块）, '^'（三角形）, 'D'（菱形）
# label 用于图例显示
plt.plot(months, sales, color='blue', linestyle='-', marker='o', label='xx产品每月销售额')

# 添加标题和坐标轴标签
plt.title('2024年每月销售额折线图', fontsize=16)
plt.xlabel('月份', fontsize=14)
plt.ylabel('销售额 (单位: 千元)', fontsize=14)

# 设置x轴刻度标签旋转角度
plt.xticks(rotation=45)  # 旋转x轴标签以防重叠

# 设置y轴范围
plt.ylim(0, 800)  # 设置y轴范围为0到800

# 添加图例
plt.legend(loc='upper left')

# 添加网格线
# grid 参数可以设置网格线的样式
# 第一个参数如果为 True 则 x轴和 y 轴都显示网格线
# alpha 参数控制网格线的透明度
plt.grid(True, linestyle='--', alpha=0.7)
# plt.grid(axis='x')  # 仅显示 x 轴网格线

# 在每个数据点上添加文本标签
for x,y in zip(months, sales):
    # y 值加20是为了让文本标签不与数据点重叠（往上平移）
    plt.text(x, y+20, str(y), ha='center', va='bottom', fontsize=10)

# 显示图表
plt.show()
```
