# 群的定义与验证



设非空集合 $  G  $ 配备一个二元运算 $  \circ : G \times G \to G  $（运算结果 $  \circ(a, b)  $ 简记为 $  a \circ b  $），若该集合与运算满足以下**四条公理**，则称 $  (G, \circ)  $ 为一个群：



1. **封闭性**：对任意 $  a, b \in G  $，有 $  a \circ b \in G  $；

2. **结合律**：对任意 $  a, b, c \in G  $，有 $  (a \circ b) \circ c = a \circ (b \circ c)  $；

3. **单位元存在性**：存在唯一元素 $  e \in G  $，使得对任意 $  a \in G  $，有 $  e \circ a = a \circ e = a  $（称 $  e  $ 为群 $  G  $ 的单位元）；

4. **逆元存在性**：对任意 $  a \in G  $，存在唯一元素 $  b \in G  $，使得 $  a \circ b = b \circ a = e  $（称 $  b  $ 为 $  a  $ 的逆元，记为 $  a^{-1}  $）。

## 补充说明

* 若群 $  (G, \circ)  $ 还满足 **交换律**（对任意 $  a, b \in G  $，有 $  a \circ b = b \circ a  $），则称其为**阿贝尔群**（或交换群）；

* 群中元素的运算优先级：括号优先，无括号时从左到右依次计算；

* 常见群例：整数集 $  \mathbb{Z}  $ 对加法构成阿贝尔群（单位元为 $  0  $，$  a  $ 的逆元为 $  -a  $），非零实数集 $  \mathbb{R}^*  $ 对乘法构成阿贝尔群（单位元为 $  1  $，$  a  $ 的逆元为 $  1/a  $）。

```python
from typing import Set, Callable, Any


# 群的核心是 “一个集合 + 一种满足特定公理的运算”
class Group:
    def __init__(self, elements: Set[Any], operation: Callable[[Any, Any], Any]):
        self.elements = elements  # 群的载体（集合）
        self.operation = operation  # 二元运算（函数）
        self.identity = self._find_identity()  # 单位元
        self._validate_group_axioms()  # 验证群公理

    def _find_identity(self) -> Any:
        """寻找单位元 e，满足对任意 a，e*a = a*e = a"""
        for e in self.elements:
            is_identity = True
            for a in self.elements:
                if self.operation(e, a) != a or self.operation(a, e) != a:
                    is_identity = False
                    break
            if is_identity:
                return e
        raise ValueError("该集合和运算不存在单位元，不构成群")

    def _has_inverse(self, a: Any) -> bool:
        """验证元素 a 是否存在逆元 b，满足 a*b = b*a = 单位元"""
        for b in self.elements:
            if (self.operation(a, b) == self.identity and
                    self.operation(b, a) == self.identity):
                return True
        return False

    def _validate_group_axioms(self) -> None:
        """验证群的4条公理：封闭性、结合律、单位元存在、逆元存在"""
        # 1. 验证封闭性：任意 a*b 的结果仍在集合中
        for a in self.elements:
            for b in self.elements:
                if self.operation(a, b) not in self.elements:
                    raise ValueError("运算不满足封闭性，不构成群")

        # 2. 验证结合律：(a*b)*c = a*(b*c)
        for a in self.elements:
            for b in self.elements:
                for c in self.elements:
                    left = self.operation(self.operation(a, b), c)
                    right = self.operation(a, self.operation(b, c))
                    if left != right:
                        raise ValueError("运算不满足结合律，不构成群")

        # 3. 单位元已在 _find_identity 中验证存在

        # 4. 验证每个元素都有逆元
        for a in self.elements:
            if not self._has_inverse(a):
                raise ValueError(f"元素 {a} 不存在逆元，不构成群")

    def is_abelian(self) -> bool:
        """判断是否为交换群（满足交换律 a*b = b*a）"""
        for a in self.elements:
            for b in self.elements:
                if self.operation(a, b) != self.operation(b, a):
                    return False
        return True
```

```python
# 定义模 3 加法运算
def add_mod3(a: int, b: int) -> int:
    return (a + b) % 3

# 构建群
elements = {0, 1, 2}
group = Group(elements, add_mod3)

print("单位元:", group.identity)  # 输出 0
print("是否为交换群:", group.is_abelian())  # 输出 True
```

# 环的定义与验证
# 环的完整定义

设非空集合 $ R $ 配备两种二元运算，分别记为加法 $ +: R \times R \to R $（运算结果记为 \( a + b \)）和乘法 $ \cdot: R \times R \to R $（运算结果记为 $ a \cdot b $或简写为 $ ab $），若集合与两种运算满足以下 **五条公理**（实际拆解为8项具体要求），则称 $ (R, +, \cdot) $ 为一个环：

## 一、环的核心公理
1. **加法封闭性**：对任意 $ a, b \in R $，有 $ a + b \in R $；  
2. **加法结合律**：对任意 $ a, b, c \in R $，有 $ (a + b) + c = a + (b + c) $；  
3. **加法交换律**：对任意 $ a, b \in R $，有 $ a + b = b + a $；  
4. **加法单位元（零元）存在性**：存在唯一元素 \( 0 \in R \)，对任意 \( a \in R \)，有 \( a + 0 = 0 + a = a \)；  
加法逆元（负元）存在性：对任意 $ a \in R $，存在唯一元素 $ -a \in R $，有 $ a + (-a) = (-a) + a = 0 $；
5. **乘法封闭性**：对任意 $ a, b \in R $，有 $ ab \in R $；
6. **乘法结合律**：对任意 $ a, b, c \in R $，有 $ (ab)c = a(bc) $；
7. **分配律**：对任意 $ a, b, c \in R $，同时满足
8. **左分配律**：$ a(b + c) = ab + ac $
9. **右分配律**：$ (b + c)a = ba + ca $


## 二、常见特殊环类型
| 环的类型 | 核心特征 | 关键说明                                                           |
|----------|----------|----------------------------------------------------------------|
| 交换环 | 乘法满足交换律 | 对任意 $a, b \in R $，有 $ ab = ba $                                |
| 含幺环 | 存在乘法单位元 | 存在 $ 1 \in R $，对任意 $ a \in R $，有 $ 1 \cdot a = a \cdot 1 = a $ |
| 整环 | 非零交换含幺环 + 无零因子 | 若 $ ab = 0 $，则必有 $ a = 0 $ 或 $ b = 0 $排除“零因子”）                 |
| 域 | 整环 + 非零元素有乘法逆元 | 对任意 $ a \in R, a \neq 0 $，存在 $ b \in R $ 使得 $ ab = ba = 1 $    |


## 三、典型示例
1. **整数环 $ (\mathbb{Z}, +, \cdot) $**：普通加法和乘法下，是交换含幺环、整环（非域，因2等元素无乘法逆元）；  
2. **矩阵环 $ (M_n(\mathbb{R}), +, \cdot) $**：n阶实矩阵的加法和乘法下，是非交换含幺环（矩阵乘法不满足交换律）；  
3. **剩余类环 $ (\mathbb{Z}_m, +, \cdot) $**：模m剩余类的加法和乘法下，m为素数时是域，m为合数时是环（非整环，存在零因子）。

```python
# 环需要 “一个集合 + 两种运算（加法和乘法）”，且加法构成交换群，乘法满足结合律和分配律。


class Ring(Group):
    def __init__(self, elements: Set[Any], add: Callable[[Any, Any], Any], mul: Callable[[Any, Any], Any]):
        # 环的加法必须是交换群，因此直接用 Group 类初始化加法结构
        super().__init__(elements, add)
        self.mul = mul  # 乘法运算
        self._validate_ring_axioms()

    def _validate_ring_axioms(self) -> None:
        """验证环的公理：乘法封闭性、结合律、分配律"""
        # 1. 乘法封闭性
        for a in self.elements:
            for b in self.elements:
                if self.mul(a, b) not in self.elements:
                    raise ValueError("乘法不满足封闭性，不构成环")

        # 2. 乘法结合律
        for a in self.elements:
            for b in self.elements:
                for c in self.elements:
                    left = self.mul(self.mul(a, b), c)
                    right = self.mul(a, self.mul(b, c))
                    if left != right:
                        raise ValueError("乘法不满足结合律，不构成环")

        # 3. 分配律：a*(b+c) = a*b + a*c 和 (b+c)*a = b*a + c*a
        for a in self.elements:
            for b in self.elements:
                for c in self.elements:
                    # 左分配律
                    left = self.mul(a, self.operation(b, c))  # a*(b+c)
                    right = self.operation(self.mul(a, b), self.mul(a, c))  # a*b + a*c
                    if left != right:
                        raise ValueError("不满足左分配律，不构成环")

                    # 右分配律
                    left = self.mul(self.operation(b, c), a)  # (b+c)*a
                    right = self.operation(self.mul(b, a), self.mul(c, a))  # b*a + c*a
                    if left != right:
                        raise ValueError("不满足右分配律，不构成环")
```

```python
def add_mod3(a: int, b: int) -> int:
    return (a + b) % 3

def mul_mod3(a: int, b: int) -> int:
    return (a * b) % 3

elements = {0, 1, 2}
ring = Ring(elements, add_mod3, mul_mod3)  # 不报错，说明构成环
```

```python
# 域是 “乘法（除 0 外）构成交换群的环”，因此需要在环的基础上验证乘法的额外公理（单位元、逆元、交换律

class Field(Ring):
    def __init__(self, elements: Set[Any], add: Callable[[Any, Any], Any], mul: Callable[[Any, Any], Any]):
        super().__init__(elements, add, mul)
        self.zero = self.identity  # 加法单位元（0）
        self.mul_identity = self._find_mul_identity()  # 乘法单位元（1）
        self._validate_field_axioms()

    def _find_mul_identity(self) -> Any:
        """寻找乘法单位元 1（≠0），满足对任意 a，1*a = a*1 = a"""
        for e in self.elements:
            if e == self.zero:
                continue  # 1 ≠ 0
            is_identity = True
            for a in self.elements:
                if self.mul(e, a) != a or self.mul(a, e) != a:
                    is_identity = False
                    break
            if is_identity:
                return e
        raise ValueError("该环不存在乘法单位元，不构成域")

    def _has_mul_inverse(self, a: Any) -> bool:
        """验证非零元素 a 是否有乘法逆元 b，满足 a*b = b*a = 1"""
        if a == self.zero:
            return False  # 0 没有乘法逆元
        for b in self.elements:
            if (self.mul(a, b) == self.mul_identity and
                    self.mul(b, a) == self.mul_identity):
                return True
        return False

    def _validate_field_axioms(self) -> None:
        """验证域的公理：乘法交换律、非零元素有逆元"""
        # 1. 乘法交换律
        for a in self.elements:
            for b in self.elements:
                if self.mul(a, b) != self.mul(b, a):
                    raise ValueError("乘法不满足交换律，不构成域")

        # 2. 所有非零元素有乘法逆元
        for a in self.elements:
            if a != self.zero and not self._has_mul_inverse(a):
                raise ValueError(f"非零元素 {a} 没有乘法逆元，不构成域")
```

```python
# 复用上面的 add_mod3 和 mul_mod3
elements = {0, 1, 2}
field = Field(elements, add_mod3, mul_mod3)  # 不报错，说明构成域

print("乘法单位元:", field.mul_identity)  # 输出 1
```
