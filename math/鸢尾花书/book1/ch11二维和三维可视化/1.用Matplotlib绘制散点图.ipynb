# 用Matplotlib绘制散点图

```python
# 导入包
import matplotlib.pyplot as plt
from sklearn.datasets import load_iris
import numpy as np
from matplotlib import rcParams # 导入rcParams以设置字体
```

```python
# 设置字体为中文
rcParams['font.sans-serif'] = ['SimHei']  # 设置字体为黑体

# 加载鸢尾花数据集
iris = load_iris()

# 提取花萼长度和花萼宽度作为变量
sepal_length = iris.data[:, 0] # 提取第一列
sepal_width = iris.data[:, 1] # 提取第二列
target = iris.target
```

```python
fig, ax = plt.subplots()

# 创建散点图
# c 参数用于指定点的颜色，这里使用虹彩色图
# cmap='rainbow' 用于指定颜色映射
plt.scatter(sepal_length, sepal_width, c=target, cmap='rainbow')

# 添加标题和轴标签
plt.title('鸢尾花萼片长度与宽度')
plt.xlabel('萼片长度 (cm)')
plt.ylabel('萼片宽度 (cm)')

# 设置横纵轴刻度
ax.set_xticks(np.arange(4, 8 + 1, step=1))
ax.set_yticks(np.arange(1, 5 + 1, step=1))

# 设定横纵轴尺度1:1
ax.axis('scaled')

# 增加刻度网格，颜色为浅灰
ax.grid(linestyle='--', linewidth=0.25, color=[0.7,0.7,0.7])

# 设置横纵轴范围
ax.set_xbound(lower = 4, upper = 8)
ax.set_ybound(lower = 1, upper = 5)

# 显示图形
plt.show()
```
