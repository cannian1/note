# 用Matplotlib生成平面等高线

```python
# 导入包
import numpy as np
import matplotlib.pyplot as plt 
```

```python
# 生成数据
x1_array = np.linspace(-3,3,121)
x2_array = np.linspace(-3,3,121)

# meshgrid 是生成网格坐标矩阵的核心函数，主要作用是将一维数组转换为二维网格
xx1, xx2 = np.meshgrid(x1_array, x2_array)
# exp 函数是计算 e 的 x 次幂
ff = xx1 * np.exp(- xx1**2 - xx2 **2)
```

`meshgrid` 可以将两个一维数组（x 轴坐标和 y 轴坐标）转换为两个二维矩阵，其中：

- 第一个矩阵的每行都是原始 x 数组的复制
- 第二个矩阵的每列都是原始 y 数组的复制

这样，两个矩阵的对应位置元素就构成了平面上所有点的坐标 (x[i,j], y[i,j])，便于批量计算。

```python
# 等高线
fig, ax = plt.subplots()

# 在轴对象 ax 上用 contour()绘制平面等高线, 返回值 CS 是一个等高线图的对象。这个对象包含了等高线图的信息，比如线条的位置、颜色等
# levels = 20 指定等高线的数量
# 'RdYlBu_r'，表示红、黄、蓝渐变的颜色映射，_r 表示翻转。
CS = ax.contour(xx1, xx2, ff, levels = 20,
                cmap = 'RdYlBu_r', linewidths = 1)

# 在图形对象 fig 上添加颜色
fig.colorbar(CS)
ax.set_xlabel('$/it{x_1}$'); ax.set_ylabel('$/it{x_2}$')
ax.set_xticks([]); ax.set_yticks([])
ax.set_xlim(xx1.min(), xx1.max())
ax.set_ylim(xx2.min(), xx2.max())
ax.grid(False)
ax.set_aspect('equal', adjustable='box')

plt.show()
```

```python
# 填充等高线
fig, ax = plt.subplots()

# contourf 函数用于绘制填充的等高线图。与 contour() 类似，但 contourf() 会在等高线之间填充颜色。
CS = ax.contourf(xx1, xx2, ff, levels = 20,
                 cmap = 'RdYlBu_r')

fig.colorbar(CS)
ax.set_xlabel('$/it{x_1}$'); ax.set_ylabel('$/it{x_2}$')
ax.set_xticks([]); ax.set_yticks([])
ax.set_xlim(xx1.min(), xx1.max())
ax.set_ylim(xx2.min(), xx2.max())
ax.grid(False)
ax.set_aspect('equal', adjustable='box')

plt.show()
```
